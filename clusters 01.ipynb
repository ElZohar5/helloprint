```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# find out clusters points

data = pd.read_csv('happyscore_income.csv')

happy = data['happyScore']
income = data['avg_income']
ineq = data['income_inequality']
import matplotlib.patches as patches

##stacking columns with np to set clusters 

income_happy = np.column_stack((income, happy))

>>> KMeans clustering data for graph_interpretation 

km_res = KMeans(n_clusters=3).fit(income_happy)
clusters = km_res.cluster_centers_

##inserting scatters

plt.scatter(income, happy)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)

##cluster's mean and std to insert ellipse function

clusters_mean = np.mean(clusters, 0)
clusters_std = np.std(clusters, 0)


#integreting patches ellipse

ellipse = patches.Ellipse([clusters_mean[0], clusters_mean[1]], clusters_std[0]*2, clusters_std[1]*2, alpha=0.25)

fig,graph = plt.subplots()

graph.scatter(income_happy[:,0], income_happy[:,1])
graph.scatter(clusters_mean[0], clusters_mean[1])
graph.add_patch(ellipse)
plt.scatter(clusters[:,0], clusters[:,1], s=1000, alpha=0.5)
```
